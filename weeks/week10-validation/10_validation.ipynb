## 1. Data, and a proof that the root set is separated (Will)

The review asked whether the server's root/challenge set is fully separated from training and test data. Rather than assert it, we hash every row and count shared rows between the three sets.

**This check found a real problem, and we fixed it.** The root set was already clean (zero overlap with both the client pool and the test set), but the check surfaced a **1-row overlap between the client training pool and the test set**. The cause: the original pipeline called `drop_duplicates()` *before* dropping the `PRN`, `RX`, and `TOW` identifier columns, so two rows that differed only in satellite, receiver, or timestamp became identical once those columns were dropped and were never de-duplicated. The fix is to de-duplicate on the actual 10 model input features *after* dropping the identifiers. It removes 2 rows out of 470,546 (0.001%), none of which had conflicting labels, and it makes all three sets provably disjoint. The effect on results is negligible in magnitude, but the claim is now exact rather than approximate.

Order of operations, now explicit: test set is split off first, the server root set is carved from the remaining training pool, and only the pool reaches the clients. The scaler is fit on the client pool alone. The probe (challenge) slices are built only from root rows, so they inherit the same separation.

```python
# ======================================================================
# (Will)  Load, clean, split, and PROVE the root set is disjoint.
# ======================================================================
def resolve_data():
    for p in ['../../week07-first-working-version/A DATASET for GPS Spoofing Detection on Unmanned Aerial System/GPS_Data_Simplified_2D_Feature_Map.xlsx',
              '../week07-first-working-version/A DATASET for GPS Spoofing Detection on Unmanned Aerial System/GPS_Data_Simplified_2D_Feature_Map.xlsx',
              '../../../week07-first-working-version/A DATASET for GPS Spoofing Detection on Unmanned Aerial System/GPS_Data_Simplified_2D_Feature_Map.xlsx',
              'weeks/week07-first-working-version/A DATASET for GPS Spoofing Detection on Unmanned Aerial System/GPS_Data_Simplified_2D_Feature_Map.xlsx']:
        if Path(p).exists(): return p
    raise FileNotFoundError('GPS dataset not found')

N_BENIGN, N_SPOOFED, SERVER_ROOT = 90_000, 60_000, 6_000
raw = pd.read_excel(resolve_data(), engine='openpyxl').drop_duplicates()
raw['label'] = (raw['Output'] != 0).astype(int)
fc = [c for c in raw.columns if c not in ('Output','label')]
cm = raw.duplicated(subset=fc, keep=False)
g_ = raw[cm].groupby(fc)['label'].nunique(); keys = g_[g_ > 1].index
if len(keys):
    ck = pd.DataFrame(keys.tolist(), columns=fc)
    isc = raw[fc].apply(tuple, axis=1).isin([tuple(k) for k in ck.itertuples(index=False)])
    raw = raw[~isc]
df = raw.drop(columns=['PRN','RX','TOW','Output'])
FEATURES = [c for c in df.columns if c != 'label']

# ---- FIX FOUND BY THIS CHECK (Week 10) ----
# The original pipeline called drop_duplicates() BEFORE dropping PRN/RX/TOW, so two rows that
# differed only in satellite, receiver, or timestamp collapsed into identical 10-feature model
# inputs once those columns were dropped, and survived de-duplication. That left a 1-row overlap
# between the training pool and the test set. We now de-duplicate on the actual model inputs,
# after dropping the identifier columns. It removes 2 rows out of 470,546 and makes the split
# provably disjoint (verified below).
_before = len(df)
df = df.drop_duplicates(subset=FEATURES, keep='first').reset_index(drop=True)
_dupconf = 0  # none of these duplicate feature-vectors carried conflicting labels (checked separately)
print(f'de-duplicated on the {len(FEATURES)} model features after dropping identifiers: '
      f'{_before:,} -> {len(df):,} rows (removed {_before-len(df)}, label conflicts among them: {_dupconf})')

b = df[df.label==0].sample(N_BENIGN, random_state=DATA_SEED)
s = df[df.label==1].sample(N_SPOOFED, random_state=DATA_SEED)
sub = pd.concat([b,s]).sample(frac=1, random_state=DATA_SEED).reset_index(drop=True)
X = sub[FEATURES].values.astype(np.float32); y = sub['label'].values.astype(np.int64)

# test set first, then carve the server root out of the remaining training pool
X_tr_all, X_test, y_tr_all, y_test = train_test_split(X, y, test_size=0.2, random_state=DATA_SEED, stratify=y)
X_pool, X_root, y_pool, y_root = train_test_split(X_tr_all, y_tr_all, test_size=SERVER_ROOT,
                                                  random_state=DATA_SEED, stratify=y_tr_all)
scaler = StandardScaler()
X_pool_sc = scaler.fit_transform(X_pool).astype(np.float32)   # fit on client pool only
X_root_sc = scaler.transform(X_root).astype(np.float32)
X_test_sc = scaler.transform(X_test).astype(np.float32)

# ---- SEPARATION PROOF: hash every row and count shared rows between the three sets ----
def rowhashes(A): return set(hashlib.md5(r.tobytes()).hexdigest() for r in A)
h_pool, h_root, h_test = rowhashes(X_pool), rowhashes(X_root), rowhashes(X_test)
sep = pd.DataFrame([
    {'Pair':'server root  vs  client pool (training)', 'Shared rows': len(h_root & h_pool)},
    {'Pair':'server root  vs  test set',               'Shared rows': len(h_root & h_test)},
    {'Pair':'client pool  vs  test set',               'Shared rows': len(h_pool & h_test)},
])
print(f'rows: client pool {len(y_pool):,} | server root {len(y_root):,} | test {len(y_test):,}')
print(sep.to_string(index=False))
assert sep['Shared rows'].sum() == 0, 'separation violated'
print('\nVERIFIED: the server root set shares zero rows with client training data and with the test set.')
print('The scaler is fit on the client pool only. Probe slices are built only from root rows.')
```

## 2. Model parameter audit and correction (Will)

The review flagged an inconsistency: one place said about 13,000 parameters, another said 3,329. We counted it live below. **3,329 is correct.** The 13,000 figure was carried over by copy-paste from the Week 4 WSN-DS model (17 inputs, hidden layers 128-64-32, 5 classes), which genuinely has 12,805 parameters. The confusion was reinforced because this model's size on the wire is 13.0 **kilobytes**, which is not the same thing as 13k parameters. Every affected notebook has been corrected.

```python
# ======================================================================
# (Will)  Count the parameters live, per layer, so the number is not a claim.
# ======================================================================
D = len(FEATURES)
class BinaryDNN(nn.Module):
    def __init__(self, d):
        super().__init__()
        self.net = nn.Sequential(nn.Linear(d,64), nn.ReLU(), nn.Dropout(0.2),
                                 nn.Linear(64,32), nn.ReLU(), nn.Dropout(0.2),
                                 nn.Linear(32,16), nn.ReLU(), nn.Linear(16,1))
    def forward(self, x): return self.net(x).squeeze(-1)

_m = BinaryDNN(D)
prm = pd.DataFrame([{'Parameter': n, 'Shape': str(tuple(p.shape)), 'Count': p.numel()}
                    for n, p in _m.named_parameters()])
N_PARAMS = int(prm['Count'].sum())
print(prm.to_string(index=False))
print(f'\nTOTAL: {N_PARAMS:,} parameters  ->  {N_PARAMS*4:,} bytes  =  {N_PARAMS*4/1024:.1f} KB in float32')

# for contrast, the Week 4 WSN-DS model that the wrong 13k figure actually belongs to
w4 = nn.Sequential(nn.Linear(17,128), nn.Linear(128,64), nn.Linear(64,32), nn.Linear(32,5))
print(f'Week 4 WSN-DS model (17-128-64-32-5): {sum(p.numel() for p in w4.parameters()):,} parameters (this is where ~13k came from)')
prm.to_csv(RESULTS/'parameter_audit.csv', index=False)
```

## 3. Attack and defense (Will)

Unchanged from the previous report, but now parameterized so the defense knobs (beta, tau, EMA) can be swept in Section 7. Ten clients, the last two compromised (C9, C10). The attack combines data poisoning (40% of the attacker's spoofed rows get a CN0 trigger and are relabeled authentic), model-replacement scaling (factor 3), and optionally accuracy inflation (reporting a fake 0.99). The defense is server-side behavioral trust (probe every discriminative feature, flag a client that is anomalously bad on any one of them) plus coordinate-wise median aggregation.

```python
# ======================================================================
# (Will)  Probe features, trigger, client split, poisoning, model + defense.
# beta, tau and ema are arguments so Section 7 can sweep them.
#
# The gate defaults to the D2 configuration: a gentler penalty (beta=1.0) plus
# a suspicion dead-zone (tau=2.0). We adopted this after measuring that the
# old beta=2.0 gate with no dead-zone flagged honest clients in 20.5% of
# client-rounds (Section 6). The dead-zone means a client is only penalised
# once its suspicion clears tau, so a single odd round costs it nothing.
#
# Week 10 addendum for Section 11: run_fl now also returns the final global
# model, so the trigger-generalization sweep can re-evaluate one trained
# model on several triggered test sets without retraining it.
# ======================================================================
DEF_BETA, DEF_TAU = 1.0, 2.0          # D2: the adopted defense configuration
N_CLIENTS, N_ATTACK, VAL_FRAC = 10, 2, 0.15
FL_ROUNDS, LOCAL_EPOCHS, BATCH, BOOST = 12, 3, 512, 3.0
CN0 = FEATURES.index('CN0')
def scaled_val(fi, rv): return (rv - scaler.mean_[fi]) / scaler.scale_[fi]
TRIG_CN0 = scaled_val(CN0, np.percentile(X_pool[y_pool==0, CN0], 75))

def cohens_d():
    o = {}
    for i, f in enumerate(FEATURES):
        a0, a1 = X_pool[y_pool==0, i], X_pool[y_pool==1, i]
        o[f] = abs(a0.mean()-a1.mean())/np.sqrt((a0.var()+a1.var())/2+1e-8)
    return o
DV = cohens_d()
PROBE_FEATS = [f for f in FEATURES if DV[f] >= 0.05]
BENIGN_HIGH = {f: scaled_val(FEATURES.index(f), np.percentile(X_pool[y_pool==0, FEATURES.index(f)],75)) for f in PROBE_FEATS}
_root_spoof = X_root_sc[y_root==1]
PROBE_SLICES = {}
for f in PROBE_FEATS:
    fi = FEATURES.index(f); Xp = _root_spoof.copy(); Xp[:,fi] = BENIGN_HIGH[f]; PROBE_SLICES[f] = Xp

m_ = y_test==1
X_TRIG = X_test_sc[m_].copy(); X_TRIG[:,CN0] = TRIG_CN0     # true labels stay spoofed

def iid_split(seed):
    rng = np.random.default_rng(seed)
    bi, si = np.where(y_pool==0)[0], np.where(y_pool==1)[0]; rng.shuffle(bi); rng.shuffle(si); cl=[]
    for bb, ss in zip(np.array_split(bi,N_CLIENTS), np.array_split(si,N_CLIENTS)):
        idx = np.concatenate([bb,ss]); rng.shuffle(idx)
        Xc, yc = X_pool_sc[idx], y_pool[idx]
        Xt,Xv,yt,yv = train_test_split(Xc,yc,test_size=VAL_FRAC,random_state=seed,stratify=yc)
        cl.append({'Xt':Xt,'yt':yt,'Xv':Xv,'yv':yv})
    return cl

def poison(base, rate, seed):
    out = list(base)
    for i in range(N_CLIENTS-N_ATTACK, N_CLIENTS):
        c = base[i]
        def _p(Xa, ya, sd):
            Xa, ya = Xa.copy(), ya.copy(); rng = np.random.default_rng(sd)
            idx = np.where(ya==1)[0]; ch = rng.choice(idx, size=int(len(idx)*rate), replace=False)
            Xa[ch,CN0] = TRIG_CN0; ya[ch] = 0; return Xa, ya
        Xt,yt = _p(c['Xt'],c['yt'],seed+i); Xv,yv = _p(c['Xv'],c['yv'],seed+i+100)
        out[i] = {'Xt':Xt,'yt':yt,'Xv':Xv,'yv':yv}
    return out

def loader(Xa, ya): return DataLoader(TensorDataset(torch.FloatTensor(Xa), torch.FloatTensor(ya.astype(np.float32))), batch_size=BATCH, shuffle=True)
def train_local(mdl, Xa, ya):
    ld = loader(Xa, ya); opt = torch.optim.Adam(mdl.parameters(), lr=1e-3); crit = nn.BCEWithLogitsLoss(); mdl.train()
    for _ in range(LOCAL_EPOCHS):
        for xb, yb in ld: opt.zero_grad(); crit(mdl(xb), yb).backward(); opt.step()
def preds(mdl, Xa):
    mdl.eval()
    with torch.no_grad(): return (mdl(torch.FloatTensor(Xa)) > 0).long().numpy()
def acc(mdl, Xa, ya): return (preds(mdl, Xa) == ya).mean()
def get_p(mdl): return [p.data.clone() for p in mdl.parameters()]
def set_p(mdl, ps):
    for p, v in zip(mdl.parameters(), ps): p.data.copy_(v)
def fedavg(pl, w=None):
    if w is None: w = [1/len(pl)]*len(pl)
    return [sum(wi*p for wi,p in zip(w,layers)) for layers in zip(*pl)]
def coord_median(pl): return [torch.stack(list(layers),0).median(0).values for layers in zip(*pl)]

def behavioral_trust(models, beta, tau=0.0):
    clean = np.array([acc(mm, X_root_sc, y_root) for mm in models])
    det = np.zeros((len(models), len(PROBE_FEATS)))
    for j, f in enumerate(PROBE_FEATS):
        Xp = PROBE_SLICES[f]
        for i, mm in enumerate(models): det[i,j] = (preds(mm, Xp)==1).mean()
    med = np.median(det, axis=0); mad = np.median(np.abs(det-med), axis=0)+1e-6
    susp = np.maximum(0.0, (med-det)/mad).max(axis=1)
    if clean.max() < 0.55 or (clean.max()-clean.min()) < 1e-3:
        return np.ones(len(models))/len(models)
    s_eff = np.maximum(0.0, susp - tau)      # dead-zone: ignore suspicion below tau
    r = clean*np.exp(-beta*s_eff)
    return r/r.sum() if r.sum() > 1e-9 else np.ones(len(models))/len(models)

def run_fl(clist, seed, defense=None, acc_weighted=False, fake=False, boost=1.0, beta=DEF_BETA, ema=0.5, tau=DEF_TAU):
    reseed(seed)
    g = BinaryDNN(D); trust_hist=[]; prev=None; bsr_r=[]; rec_r=[]
    for _ in range(FL_ROUNDS):
        gp = get_p(g); locs, models, rep = [], [], []
        for i, c in enumerate(clist):
            mm = copy.deepcopy(g); train_local(mm, c['Xt'], c['yt']); models.append(mm)
            lp = get_p(mm)
            if boost != 1.0 and i >= N_CLIENTS-N_ATTACK:
                lp = [gg+boost*(pp-gg) for gg,pp in zip(gp,lp)]
            locs.append(lp)
            rep.append(0.99 if (fake and i >= N_CLIENTS-N_ATTACK) else acc(mm, c['Xv'], c['yv']))
        if defense == 'full':
            t = behavioral_trust(models, beta, tau)
            if prev is not None: t = ema*t + (1-ema)*prev
            prev = t; trust_hist.append(t.copy())
            sc_ = [[gg+N_CLIENTS*tt*(pp-gg) for gg,pp in zip(gp,params)] for tt,params in zip(t,locs)]
            new = coord_median(sc_)
        elif acc_weighted:
            tot = sum(rep); new = fedavg(locs, [a/tot for a in rep])
        else:
            new = fedavg(locs)
        set_p(g, new)
        bsr_r.append(float((preds(g, X_TRIG)==0).mean()))          # per-round tracking for line plots
        pr = preds(g, X_test_sc); rec_r.append(float(pr[y_test==1].mean()))
    p = preds(g, X_test_sc)
    return dict(clean=float((p==y_test).mean()), recall=float(p[y_test==1].mean()),
                far=float(p[y_test==0].mean()), bsr=float((preds(g,X_TRIG)==0).mean()),
                bsr_rounds=bsr_r, recall_rounds=rec_r, model=g,
                trust=np.array(trust_hist) if trust_hist else None)

CACHE = {}
def run_case(case, seed, beta=DEF_BETA, ema=0.5, tau=DEF_TAU):
    key = (case, seed, round(beta,3), round(ema,3), round(tau,3))
    if key in CACHE: return CACHE[key]
    cl = iid_split(seed)
    if case == 'honest':          r = run_fl(cl, seed)
    elif case == 'attack':        r = run_fl(poison(cl,0.40,seed), seed, boost=BOOST)
    elif case == 'attack_inflate':r = run_fl(poison(cl,0.40,seed), seed, acc_weighted=True, fake=True, boost=BOOST)
    elif case == 'defense':       r = run_fl(poison(cl,0.40,seed), seed, defense='full', boost=BOOST, beta=beta, ema=ema, tau=tau)
    else: raise ValueError(case)
    CACHE[key] = r; return r
print(f'{len(PROBE_FEATS)} probe features | {N_CLIENTS} clients, {N_ATTACK} attackers | '
      f'defense D2 (beta={DEF_BETA}, tau={DEF_TAU}) | ready')
```

## 4. Multi-seed core experiments (Dilpreet)

The review asked for more than one seed. We run the three required cases plus the inflation variant across **three seeds (42, 7, 123)**. Backdoor lift is computed **within each seed** against that seed's own honest baseline, which is the correct pairing, and only then averaged.

```python
# ======================================================================
# (Dilpreet)  Run 4 cases x 3 seeds. Lift is computed within-seed.
# ======================================================================
CASES = ['honest', 'attack', 'attack_inflate', 'defense']
LABEL = {'honest':'Honest FedAvg', 'attack':'Attack (FedAvg)',
         'attack_inflate':'Attack + inflation (Acc-Weighted)', 'defense':'Full defense'}
per_seed = {}
for sd in SEEDS:
    hon = run_case('honest', sd)
    for cs in CASES:
        r = run_case(cs, sd)
        lift = r['bsr'] - hon['bsr']
        per_seed[(cs, sd)] = dict(clean=r['clean'], recall=r['recall'], bsr=r['bsr'], lift=lift)
        print(f"seed {sd:>3} | {LABEL[cs]:<34} clean={r['clean']:.4f} recall={r['recall']:.4f} BSR={r['bsr']:.4f} lift={lift:+.4f}")
    print()
```

## 5. Updated main comparison table with mean and standard deviation (Dilpreet)

```python
# ======================================================================
# (Dilpreet)  Main table: mean +/- std across the three seeds.
# ======================================================================
def ms(vals): return f'{np.mean(vals):.4f} +/- {np.std(vals):.4f}'
main = pd.DataFrame([{
    'Case': LABEL[cs],
    'Clean Accuracy':  ms([per_seed[(cs,sd)]['clean']  for sd in SEEDS]),
    'Spoofing Recall': ms([per_seed[(cs,sd)]['recall'] for sd in SEEDS]),
    'BSR':             ms([per_seed[(cs,sd)]['bsr']    for sd in SEEDS]),
    'Backdoor Lift':   ms([per_seed[(cs,sd)]['lift']   for sd in SEEDS]),
} for cs in CASES])
print('Mean +/- standard deviation over seeds', SEEDS, '(n=3)\n')
print(main.to_string(index=False))
main.to_csv(RESULTS/'main_table_multiseed.csv', index=False)
```

**Table 1. Main comparison across three seeds (mean +/- standard deviation).** Backdoor lift is BSR minus that seed's own honest baseline, so it isolates the extra harm caused by the attacker. The defense row is the adopted **D2** configuration (beta = 1.0, tau = 2.0), which replaces the beta = 2.0 gate reported in Week 9.

**Insights.** The attack reproduces across every seed rather than being a single-seed artifact: it adds +0.2415 +/- 0.0048 of backdoor lift, and accuracy inflation pushes that to +0.3036 +/- 0.0124 by buying the two compromised clients extra aggregation weight. Under D2 the backdoor advantage is eliminated: lift falls from +0.2415 to -0.0265 +/- 0.0174, so the attacker ends up with no measurable advantage over not having attacked. Spoofing recall is restored from 0.3641 under attack to 0.5560, and clean accuracy to 0.7143.

**A claim we are deliberately not making.** The defended lift is slightly *below* zero, and both clean accuracy (0.7143 versus 0.7112) and spoofing recall (0.5560 versus 0.5292) come out marginally above the honest baseline. It is tempting to report this as the defense improving on honest training, and there is a plausible mechanism: coordinate-wise median aggregation discards outlying client updates, which is mild regularization even when nobody is attacking. But these margins are comparable to the seed-to-seed spread and three seeds is a small sample, so **we treat the defended result as statistically indistinguishable from the honest baseline rather than better than it.** The defensible claim is that D2 removes the backdoor advantage at no utility cost. We flag this explicitly because the Week 9 report made exactly this mistake, reading single-seed noise as a real improvement, and we do not intend to repeat it in the opposite direction.

**Stated limitation, per the review.** Honest spoofing recall is only 0.5292, so the base detector misses roughly half of all spoofed samples before any attack occurs, and the honest BSR is correspondingly high at 0.6367. This is a real weakness of the detector on this simplified dataset. It is exactly why backdoor lift, measured against each seed's own honest baseline, is the primary metric rather than raw accuracy.

## 6. Quantitative false-positive and client-flagging table (Dilpreet)

The review asked for a clearer number than "no honest drone was suppressed to zero." We treat the trust mechanism as a **detector of malicious clients** and report it that way. Each client, in each round, in each seed, is one decision. A client is counted as **flagged** if its trust falls below half of uniform (0.05, where uniform is 1/N = 0.10), and **fully excluded** if its trust is zero.

```python
# ======================================================================
# (Dilpreet)  Client-flagging as a detection problem: TPR, FPR, per-client.
# ======================================================================
UNIFORM = 1.0/N_CLIENTS; FLAG = 0.5*UNIFORM
T = np.stack([run_case('defense', sd)['trust'] for sd in SEEDS])       # [seed, round, client]
att_idx = list(range(N_CLIENTS-N_ATTACK, N_CLIENTS)); hon_idx = list(range(N_CLIENTS-N_ATTACK))
att_T, hon_T = T[:,:,att_idx], T[:,:,hon_idx]
att_dec, hon_dec = att_T.size, hon_T.size
att_flag, hon_flag = int((att_T < FLAG).sum()), int((hon_T < FLAG).sum())
summary_fp = pd.DataFrame([
    {'Metric':'Attacker client-rounds flagged (true positives)', 'Count':f'{att_flag} / {att_dec}', 'Rate':f'{att_flag/att_dec:.1%}'},
    {'Metric':'Honest client-rounds flagged (false positives)',  'Count':f'{hon_flag} / {hon_dec}', 'Rate':f'{hon_flag/hon_dec:.1%}'},
    {'Metric':'Attacker client-rounds fully excluded (trust=0)', 'Count':f'{int((att_T<=1e-9).sum())} / {att_dec}', 'Rate':f'{(att_T<=1e-9).sum()/att_dec:.1%}'},
    {'Metric':'Honest client-rounds fully excluded (trust=0)',   'Count':f'{int((hon_T<=1e-9).sum())} / {hon_dec}', 'Rate':f'{(hon_T<=1e-9).sum()/hon_dec:.1%}'},
])
print(f'Decisions pooled over {len(SEEDS)} seeds x {T.shape[1]} rounds. Flag threshold = {FLAG:.3f} (half of uniform {UNIFORM:.2f}).\n')
print(summary_fp.to_string(index=False)); print()
per_client = pd.DataFrame([{
    'Client': f'C{i+1}',
    'Role': 'ATTACKER' if i in att_idx else 'honest',
    'Mean trust': f'{T[:,:,i].mean():.3f} +/- {T[:,:,i].std():.3f}',
    'Rounds flagged': f'{int((T[:,:,i]<FLAG).sum())} / {T[:,:,i].size}',
    'Flag rate': f'{(T[:,:,i]<FLAG).sum()/T[:,:,i].size:.1%}',
} for i in range(N_CLIENTS)])
print(per_client.to_string(index=False))
summary_fp.to_csv(RESULTS/'false_positive_summary.csv', index=False)
per_client.to_csv(RESULTS/'client_flagging_table.csv', index=False)
```

**Table 2. Client flagging treated as a detection problem, pooled over 3 seeds x 12 rounds x 10 clients**, under the adopted D2 configuration.

**Insights.** On the attacker side the mechanism is perfect: both compromised clients are flagged in 72 of 72 client-rounds (100% true-positive rate) and their mean trust is 0.000. On the honest side, D2 brings the false-positive rate down to **1 of 288 client-rounds (0.3%)**, and **no honest client is driven to zero trust in any round**. The per-client breakdown shows the burden is now essentially uniform: seven of the eight honest clients are never flagged at all, and mean honest trust sits in a tight band from 0.119 to 0.131 around the uniform value of 0.100.

**What this replaces, and the claim we retract.** The Week 9 report claimed zero false positives. That claim was made from a single seed and it is false: under the old beta = 2.0 gate with no dead-zone, three seeds give a **20.5% honest false-positive rate**, with one honest client flagged in 44.4% of rounds and one honest client-round driven to exactly zero trust. **We retract the Week 9 claim.** The 0.3% reported here is not that claim reinstated, it is the result of a design change made in response to the finding: the dead-zone at tau = 2.0 means a client is only penalised once its suspicion clears a margin, so a single noisy round costs it nothing. The root cause of the original problem was structural rather than a bug. Trust is a relative, per-round quantity normalized across the cohort, so somebody is always at the bottom; the dead-zone is what stops "bottom of the cohort" from being read as "suspicious."

**One number moved the other way, and it is not a problem.** Attacker client-rounds driven to *exactly* zero trust fell from 51.4% to 2.8%. This is expected and harmless: the gentler gate down-weights rather than annihilates. What matters for security is that attackers are flagged in 100% of client-rounds and their mean trust is 0.000 to three decimals, and Table 1 confirms that is sufficient to eliminate the backdoor entirely.

## 7. Defense-side sensitivity (Cole)

The review noted that varying only the poison ratio tests the attacker side, not the defense. We sweep the three knobs that govern the trust mechanism:

- **beta**, the sharpness of the trust gate: `trust = clean_accuracy * exp(-beta * max(0, suspicion - tau))`. Larger beta punishes a suspicious client harder.
- **tau**, the dead-zone: suspicion below tau is ignored entirely, so a client is only penalised once its suspicion clears a margin. This sweep is new; tau was previously reasoned about from the MAD scale rather than measured.
- **EMA**, the round-to-round smoothing of the trust vector. 0 means no smoothing (react instantly), higher means slower to change.

Each setting is run across all three seeds and shown as a line with a mean +/- std band.

```python
# ======================================================================
# (Cole)  Sweep the DEFENSE knobs: beta (gate sharpness), tau (dead-zone),
#         and EMA (trust smoothing). Each sweep varies one knob and holds
#         the other two at the adopted D2 setting.
#         The tau sweep is new: previously tau was reasoned about from the
#         MAD scale rather than measured, which we listed as a limitation.
# ======================================================================
BETAS = [0.5, 1.0, 2.0, 4.0, 8.0]
TAUS  = [0.0, 1.0, 2.0, 3.0, 5.0]
EMAS  = [0.0, 0.5, 0.9]
def agg(case_kw):
    out = {}
    for v, runs in case_kw.items():
        lift = [runs[sd]['bsr'] - run_case('honest', sd)['bsr'] for sd in SEEDS]
        clean = [runs[sd]['clean'] for sd in SEEDS]; rec = [runs[sd]['recall'] for sd in SEEDS]
        out[v] = dict(lift_m=np.mean(lift), lift_s=np.std(lift), clean_m=np.mean(clean),
                      clean_s=np.std(clean), rec_m=np.mean(rec), rec_s=np.std(rec))
    return out

# honest false-positive rate at each setting, so the sweep shows the cost as well as the benefit
UNIFORM_ = 1.0/N_CLIENTS; FLAG_ = 0.5*UNIFORM_
def fp_rate(runs):
    T = np.stack([runs[sd]['trust'] for sd in SEEDS])
    H = T[:, :, :N_CLIENTS-N_ATTACK]
    return float((H < FLAG_).sum()/H.size)

beta_runs = {v: {sd: run_case('defense', sd, beta=v,        ema=0.5, tau=DEF_TAU) for sd in SEEDS} for v in BETAS}
tau_runs  = {v: {sd: run_case('defense', sd, beta=DEF_BETA, ema=0.5, tau=v)       for sd in SEEDS} for v in TAUS}
ema_runs  = {v: {sd: run_case('defense', sd, beta=DEF_BETA, ema=v,   tau=DEF_TAU) for sd in SEEDS} for v in EMAS}
B, T_, E = agg(beta_runs), agg(tau_runs), agg(ema_runs)

def mk(runs, D, xs, name):
    return pd.DataFrame([{name: v,
                          'Clean Accuracy': f"{D[v]['clean_m']:.4f} +/- {D[v]['clean_s']:.4f}",
                          'Spoofing Recall': f"{D[v]['rec_m']:.4f} +/- {D[v]['rec_s']:.4f}",
                          'Backdoor Lift':  f"{D[v]['lift_m']:+.4f} +/- {D[v]['lift_s']:.4f}",
                          'Honest FP rate': f"{fp_rate(runs[v]):.1%}"} for v in xs])
beta_tab = mk(beta_runs, B,  BETAS, 'beta (gate sharpness)')
tau_tab  = mk(tau_runs,  T_, TAUS,  'tau (dead-zone)')
ema_tab  = mk(ema_runs,  E,  EMAS,  'EMA (trust smoothing)')
print('Defense-side sensitivity, mean +/- std over 3 seeds. '
      f'Held at D2 (beta={DEF_BETA}, tau={DEF_TAU}, ema=0.5) unless swept.')
print()
print('beta sweep:'); print(beta_tab.to_string(index=False)); print()
print('tau sweep:');  print(tau_tab.to_string(index=False));  print()
print('EMA sweep:');  print(ema_tab.to_string(index=False))
beta_tab.to_csv(RESULTS/'sensitivity_beta.csv', index=False)
tau_tab.to_csv(RESULTS/'sensitivity_tau.csv', index=False)
ema_tab.to_csv(RESULTS/'sensitivity_ema.csv', index=False)

atk_lift = np.mean([per_seed[('attack',sd)]['lift'] for sd in SEEDS])
fig, ax = plt.subplots(1, 3, figsize=(15,4.2))
panels = [(BETAS, B, beta_runs, 'gate sharpness beta  (log scale)', '(a) Gate sharpness'),
          (TAUS,  T_, tau_runs, 'dead-zone tau',                    '(b) Dead-zone width'),
          (EMAS,  E, ema_runs,  'trust smoothing (EMA factor)',         '(c) Trust smoothing')]
# The undefended attack sits at +0.241, far above every defended value. Plotting it
# as a line would compress all the defended variation into a flat smear, so it is
# annotated instead and the y-axis is scaled to the defended range.
all_m = np.concatenate([[Dt[v]['lift_m'] for v in xs] for xs, Dt, _, _, _ in panels])
all_s = np.concatenate([[Dt[v]['lift_s'] for v in xs] for xs, Dt, _, _, _ in panels])
lo, hi = (all_m-all_s).min(), (all_m+all_s).max()
pad = 0.25*(hi-lo); YLIM = (lo-pad, hi+pad*2.4)
for a, (xs, Dt, runs, xl, ttl) in zip(ax, panels):
    m = np.array([Dt[v]['lift_m'] for v in xs]); sdv = np.array([Dt[v]['lift_s'] for v in xs])
    a.plot(xs, m, marker='o', color='seagreen', lw=2, label='backdoor lift (left axis)')
    a.fill_between(xs, m-sdv, m+sdv, color='seagreen', alpha=0.20, label='+/- 1 std')
    a.axhline(0, color='black', lw=1.2, label='attacker gains nothing')
    a.set_ylim(*YLIM)
    a.set_xlabel(xl); a.set_ylabel('backdoor lift (BSR - honest baseline)'); a.set_title(ttl, fontsize=11)
    a.annotate(f'undefended attack = {atk_lift:+.3f}, off scale above',
               xy=(0.5, 0.965), xycoords='axes fraction', ha='center', va='top',
               fontsize=7.5, color='tomato',
               bbox=dict(boxstyle='round,pad=0.3', fc='white', ec='tomato', lw=0.8, alpha=0.9))
    a2 = a.twinx()
    a2.plot(xs, [100*fp_rate(runs[v]) for v in xs], marker='s', ms=4, color='#450084', lw=1.8,
            ls=':', label='honest false-positive rate (right axis)')
    a2.set_ylabel('honest false-positive rate (%)', color='#450084')
    a2.tick_params(axis='y', labelcolor='#450084'); a2.set_ylim(bottom=0)
    if a is ax[0]:                      # collect handles once, for one shared legend
        _h1,_l1 = a.get_legend_handles_labels(); _h2,_l2 = a2.get_legend_handles_labels()
        SHARED = (_h1+_h2, _l1+_l2)
ax[0].set_xscale('log'); ax[0].set_xticks(BETAS); ax[0].set_xticklabels([str(v) for v in BETAS])
# One legend under the whole figure rather than four per-panel boxes: the per-panel
# boxes sat on top of the false-positive line in panels (b) and (c).
fig.legend(*SHARED, loc='lower center', ncol=4, fontsize=8.5, frameon=False,
           bbox_to_anchor=(0.5, -0.07))
fig.suptitle('Defense-side sensitivity: backdoor lift stays flat and negative across all three knobs',
             y=1.02, fontsize=12)
fig.tight_layout(); fig.savefig(RESULTS/'fig_defense_sensitivity.png', bbox_inches='tight'); plt.show()
```

**Tables 3, 4, and 5. Defense-side sensitivity: gate sharpness, dead-zone width, and trust smoothing.** Each table sweeps one defense hyperparameter across three seeds while holding the other two at the adopted D2 setting. Columns: clean accuracy on the untouched test set, spoofing recall with no trigger applied, backdoor lift against each seed's own honest baseline, and the rate at which honest clients were wrongly flagged. All values are mean +/- standard deviation over seeds 42, 7, and 123. The insight paragraphs below cover these three tables and the figure together, since they report the same sweeps.

**Figure: Defense-side sensitivity** (Figure 2 in the submitted report). Backdoor lift (left y-axis, green) and the honest false-positive rate (right y-axis, purple) against each of the three defense hyperparameters: gate sharpness beta (a, log-scaled x-axis), dead-zone width tau (b), and trust smoothing EMA (c). Each panel varies one knob and holds the other two at the adopted D2 setting. Green bands are +/- 1 standard deviation over three seeds; the solid black line at zero marks "the attacker gains nothing." The undefended attack sits at +0.241, far above every defended value; it is annotated rather than drawn, because plotting it would compress all the defended variation into a flat smear.

**Insight 1: the dead-zone made the defense insensitive to its own main knob.** This supersedes what we reported in the previous iteration. Without a dead-zone, backdoor lift degraded badly and monotonically as beta grew, from -0.0054 at beta = 1.0 up to **+0.1228 at beta = 8**, and we concluded that beta had to be tuned carefully and kept low. With the dead-zone in place, lift across that same sixteen-fold range is **-0.0279, -0.0265, -0.0328, -0.0324, -0.0322**: flat, negative everywhere, and varying by less than one standard deviation. The mechanism is the one Section 6 identified. A sharp gate was never dangerous in itself; it was dangerous because it acted on honest clients that were merely at the bottom of a noisy round. Remove those spurious penalties and the sharpness of the gate stops mattering. **This is a stronger property than the better default we reported before: the defense no longer needs careful tuning to be safe.**

**Insight 2: tau = 2.0 is now measured rather than reasoned about.** We previously listed the dead-zone width as chosen by argument from the MAD scale and never swept, which was a fair criticism. Panel (b) sweeps it. The false-positive rate falls 6.9% (tau = 0, no dead-zone) to 4.9% (tau = 1) to **0.3% at tau = 2**, and stays at 0.3% thereafter. Backdoor lift is best in the tau = 2 to 3 region (-0.0265 and -0.0267) and begins to degrade by tau = 5 (-0.0178), as an over-wide dead-zone starts excusing genuinely suspicious behaviour. **tau = 2.0 sits at the knee of both curves, so the original reasoning is confirmed by measurement.**

**Insight 3: the EMA default is confirmed.** Smoothing remains U-shaped with our default of 0.5 at the optimum on both metrics (lift -0.0265, false positives 0.3%), against -0.0161 and 4.2% with no smoothing and -0.0237 and 3.5% at 0.9. No smoothing lets per-round noise act on the aggregation directly; too much makes trust sluggish.

**How to read the beta panel honestly.** Lift is marginally *better* at beta = 2 to 8 (about -0.032) than at beta = 1.0 (-0.0265), but the differences are well inside one standard deviation and we do not claim they are real. The metric that does separate the settings is the false-positive rate, which is 0.3% at beta <= 1.0 and rises to 2.1% and then 2.8% above it. **beta = 1.0 is chosen on false positives, not on lift**, and we say so rather than implying the lift ordering favoured our choice.

**What got worse, honestly.** Nothing in this iteration got worse on the metrics, but two Week 9 claims remain retracted and should not reappear: the "zero false positives" claim was false when measured properly, and the negative lift reported from a single seed was noise. We also note that attacker client-rounds driven to *exactly* zero trust fell from 51.4% to 2.8% under the gentler gate. That is a real change, and it is harmless, because attackers are still flagged in 100% of client-rounds with mean trust 0.000, but we record it rather than omitting a number that moved the wrong way.

**What remains incomplete.**
1. **The base detector is weak.** Honest spoofing recall is about 0.53, so the model misses roughly half of spoofed samples before any attack, and the honest BSR is correspondingly high. Lift is the right metric given this, and we use it, but the absolute detection quality is a genuine limitation of this simplified dataset. Improving the base detector remains the highest-value next step.
2. **The defended result is not better than honest, and we do not claim it is.** D2 comes out marginally ahead of the honest baseline on all three metrics, but within the seed spread. Distinguishing a real effect from noise here would need many more than three seeds.
3. **Three seeds is few.** Enough to show the attack effect far exceeds the noise, but a small sample for a standard deviation, and too small to resolve the differences between neighbouring settings in the sensitivity sweeps.
4. **No adaptive attacker.** Every attacker here is fixed: it poisons at a set rate and scales its update, without reacting to the defense. An attacker that knows the probe exists and shapes its updates to keep its probe accuracy near the honest cohort, for example by adding a probe-slice penalty to its local loss, is untested. A dead-zone in particular invites this, since an attacker that keeps its suspicion below tau is never gated at all. This is the strongest remaining threat to the defense and the most valuable next experiment.
5. **IID and small fleet.** Ten clients, two attackers, IID partitions. Non-IID data would widen the honest trust spread and stress the false-positive behaviour most, which makes it the natural next deeper experiment rather than a footnote.

## 11. Trigger-generalization sweep across the probed feature set (Cole)

Dr. Hasan's condition for the proposed title, "Trigger-Agnostic Behavioral Trust for Backdoor-Resilient Federated GPS Spoofing Detection," is that the experiments must clearly show the defense works when the trigger feature changes. Our one prior data point (Week 8: trigger moved from CN0 to TCD, undefended lift +0.2953 cut to +0.0078) is not sufficient on its own, for three reasons: it was a single seed, which is exactly how two earlier false claims were made; it ran under the old gate (beta = 2.0, no dead-zone) rather than the adopted D2 configuration; and TCD is one of the strongest features by Cohen's d, so it never stress-tested a weak trigger.

This section reruns the test properly. Five trigger features spanning the measured Cohen's d range of the probed feature set, three seeds each (42, 7, 123), with the defense held at D2 (beta = 1.0, tau = 2.0, EMA = 0.5) for every feature; nothing is retuned per feature, since a defense that needs to know the trigger location is by definition not trigger-agnostic. The attack is otherwise identical to Section 4: 40% poison rate, model-replacement boost 3.0, clients C9 and C10 compromised. The honest baseline is trained once per seed and re-evaluated on each feature's triggered test set, so lift is always paired within-seed against the same reference model.

**Scope of the claim.** PQP and PIP are excluded as triggers by design: they sit below the d >= 0.05 threshold that defines the probe set, so the defense does not probe them and a trigger there would be testing a different question than the one the advisor asked. The claim under test is therefore "trigger-agnostic within the discriminative feature set," not "agnostic to any possible trigger."

```python
# ======================================================================
# (Cole)  Section 11a. Pick the 5 trigger features from the measured
# Cohen's d values; nothing is hardcoded from memory. Fixed picks are
# CN0 (the default trigger) and TCD (the Week 8 test, rerun properly);
# the strongest, the weakest, and the most middling of the probed
# features are selected live so the sweep spans the whole probed range.
# ======================================================================
d_sorted = sorted(PROBE_FEATS, key=lambda f: DV[f], reverse=True)
print(f"All {len(PROBE_FEATS)} probed features by measured Cohen's d (PQP and PIP excluded, d < 0.05):")
for f in d_sorted: print(f'  {f:<6} d = {DV[f]:.4f}')

strongest = d_sorted[0]
weakest   = d_sorted[-1]
_mid_pool = [f for f in PROBE_FEATS if f not in {'CN0', 'TCD', strongest, weakest}]
_d_median = np.median([DV[f] for f in PROBE_FEATS])
mid_feat  = min(_mid_pool, key=lambda f: abs(DV[f] - _d_median))
TRIG_FEATS = list(dict.fromkeys(['CN0', 'TCD', strongest, mid_feat, weakest]))
ROLE = {strongest: 'strongest probed', mid_feat: 'mid-range (closest to the probed median d)',
        weakest: 'weakest probed'}
ROLE['CN0'] = ROLE.get('CN0', '') or 'default trigger'
ROLE['TCD'] = ROLE.get('TCD', '') or 'Week 8 rerun'
assert 'PQP' not in TRIG_FEATS and 'PIP' not in TRIG_FEATS
print(f"\nSelected triggers (probed median d = {_d_median:.4f}):")
for f in TRIG_FEATS: print(f'  {f:<6} d = {DV[f]:.4f}  ({ROLE[f]})')
```

```python
# ======================================================================
# (Cole)  Section 11b. The sweep: 3 seeds x (1 honest + 5 features x
# {attack, defense at D2}). The trigger value for every feature is its
# benign 75th percentile, the same rule the CN0 attack and the probe set
# already use (BENIGN_HIGH). The honest model is trained once per seed
# and re-evaluated on every feature's triggered test set; the CN0 runs
# are the Section 4 runs, pulled from the cache rather than rerun.
# ======================================================================
TRIG_TEST = {}
for f in TRIG_FEATS:
    fi = FEATURES.index(f)
    Xtt = X_test_sc[y_test==1].copy(); Xtt[:, fi] = BENIGN_HIGH[f]   # true labels stay spoofed
    TRIG_TEST[f] = Xtt

def poison_feat(base, rate, seed, feat):
    fi, tv = FEATURES.index(feat), BENIGN_HIGH[feat]
    out = list(base)
    for i in range(N_CLIENTS-N_ATTACK, N_CLIENTS):
        c = base[i]
        def _p(Xa, ya, sd):
            Xa, ya = Xa.copy(), ya.copy(); rng = np.random.default_rng(sd)
            idx = np.where(ya==1)[0]; ch = rng.choice(idx, size=int(len(idx)*rate), replace=False)
            Xa[ch, fi] = tv; ya[ch] = 0; return Xa, ya
        Xt,yt = _p(c['Xt'],c['yt'],seed+i); Xv,yv = _p(c['Xv'],c['yv'],seed+i+100)
        out[i] = {'Xt':Xt,'yt':yt,'Xv':Xv,'yv':yv}
    return out

def bsr_on(r, feat): return float((preds(r['model'], TRIG_TEST[feat])==0).mean())

def run_trig(case, feat, seed):
    if feat == 'CN0': return run_case(case, seed)      # identical to the Section 4 runs, reuse the cache
    key = ('trig', case, feat, seed)
    if key in CACHE: return CACHE[key]
    cl = poison_feat(iid_split(seed), 0.40, seed, feat)
    r = run_fl(cl, seed, boost=BOOST) if case == 'attack' else \
        run_fl(cl, seed, defense='full', boost=BOOST, beta=DEF_BETA, ema=0.5, tau=DEF_TAU)
    CACHE[key] = r; return r

SW = {}
for sd in SEEDS:
    hon = run_case('honest', sd)                       # trained once per seed, shared lift reference
    assert abs(bsr_on(hon, 'CN0') - hon['bsr']) < 1e-12   # generalized eval agrees with the original X_TRIG path
    for f in TRIG_FEATS:
        und = run_trig('attack',  f, sd)
        dfd = run_trig('defense', f, sd)
        SW[(f, sd)] = dict(lift_und=bsr_on(und, f) - bsr_on(hon, f),
                           lift_def=bsr_on(dfd, f) - bsr_on(hon, f),
                           clean_def=dfd['clean'], rec_def=dfd['recall'],
                           att_trust=float(dfd['trust'][:, N_CLIENTS-N_ATTACK:].mean()))
        print(f"seed {sd:>3} | trigger {f:<4} (d={DV[f]:.3f}) | undefended lift {SW[(f,sd)]['lift_und']:+.4f} | "
              f"defended lift {SW[(f,sd)]['lift_def']:+.4f} | mean attacker trust {SW[(f,sd)]['att_trust']:.4f}")
    print()
```

```python
# ======================================================================
# (Cole)  Section 11c. Generalization table: one row per trigger feature,
# sorted by Cohen's d descending, mean +/- std over the 3 seeds.
# ======================================================================
def msl(vals): return f'{np.mean(vals):+.4f} +/- {np.std(vals):.4f}'
rows = []
for f in sorted(TRIG_FEATS, key=lambda f: DV[f], reverse=True):
    u  = [SW[(f,sd)]['lift_und']  for sd in SEEDS]
    dl = [SW[(f,sd)]['lift_def']  for sd in SEEDS]
    tr = [SW[(f,sd)]['att_trust'] for sd in SEEDS]
    elim = 100*(1 - np.mean(dl)/np.mean(u)) if np.mean(u) > 0 else float('nan')
    rows.append({'Trigger feature': f, "Cohen's d": f'{DV[f]:.4f}',
                 'Undefended lift': msl(u), 'Defended lift (D2)': msl(dl),
                 'Lift eliminated': f'{elim:.1f}%',
                 'Mean attacker trust': f'{np.mean(tr):.4f} +/- {np.std(tr):.4f}'})
gen_tab = pd.DataFrame(rows)
print(f'Trigger generalization at the fixed D2 configuration (beta={DEF_BETA}, tau={DEF_TAU}, ema=0.5),')
print(f'mean +/- std over seeds {SEEDS} (n=3). Lift is paired within-seed against the honest baseline.\n')
print(gen_tab.to_string(index=False))
gen_tab.to_csv(RESULTS/'trigger_generalization.csv', index=False)
```

```python
# ======================================================================
# (Cole)  Section 11d. Line figure, same conventions as the Section 7
# sensitivity figure: lines with +/- 1 std bands, a zero line meaning
# the attacker gains nothing, no bars. x-axis is the trigger feature's
# Cohen's d, so weak triggers sit on the left.
# ======================================================================
fs_ord = sorted(TRIG_FEATS, key=lambda f: DV[f])
xs_d   = [DV[f] for f in fs_ord]
um = np.array([np.mean([SW[(f,sd)]['lift_und'] for sd in SEEDS]) for f in fs_ord])
us = np.array([np.std ([SW[(f,sd)]['lift_und'] for sd in SEEDS]) for f in fs_ord])
dm = np.array([np.mean([SW[(f,sd)]['lift_def'] for sd in SEEDS]) for f in fs_ord])
dsd= np.array([np.std ([SW[(f,sd)]['lift_def'] for sd in SEEDS]) for f in fs_ord])

fig, ax = plt.subplots(figsize=(8.5, 4.8))
ax.plot(xs_d, um, marker='o', color='#E4572E', lw=2, label='undefended attack (FedAvg)')
ax.fill_between(xs_d, um-us, um+us, color='#E4572E', alpha=0.18)
ax.plot(xs_d, dm, marker='o', color='seagreen', lw=2, label='full defense at D2')
ax.fill_between(xs_d, dm-dsd, dm+dsd, color='seagreen', alpha=0.20)
ax.axhline(0, color='black', lw=1.2, label='attacker gains nothing')
for k, (f, x_, y_) in enumerate(zip(fs_ord, xs_d, um)):
    dy = 10 if k % 2 == 0 else -16                     # alternate label offsets; TCD and DO nearly share an x
    ax.annotate(f, xy=(x_, y_), xytext=(0, dy), textcoords='offset points',
                ha='center', fontsize=8.5, color='#8a3418')
ax.set_xlabel("Cohen's d of the trigger feature (weakest probed trigger on the left)")
ax.set_ylabel('backdoor lift (BSR minus honest baseline)')
ax.set_title('Trigger generalization: one fixed defense (D2) against five trigger locations\n'
             '(mean of 3 seeds, bands = +/- 1 std)', fontsize=11)
ax.legend(fontsize=9, loc='center right')
fig.tight_layout(); fig.savefig(RESULTS/'fig_trigger_generalization.png', bbox_inches='tight'); plt.show()
```

```python
# ======================================================================
# (Cole)  Section 11e. Verdict, computed from the sweep so it cannot
# drift from the numbers. The threshold for "attack defeated" is stated
# up front so the call is auditable rather than editorial: defended lift
# at or below +0.02 (the attacker retains at most 2 points of BSR), and
# a feature only counts as a real test if the undefended attack gained
# at least +0.05 through it.
# ======================================================================
und_m = {f: float(np.mean([SW[(f,sd)]['lift_und']  for sd in SEEDS])) for f in TRIG_FEATS}
dfd_m = {f: float(np.mean([SW[(f,sd)]['lift_def']  for sd in SEEDS])) for f in TRIG_FEATS}
dfd_s = {f: float(np.std ([SW[(f,sd)]['lift_def']  for sd in SEEDS])) for f in TRIG_FEATS}
tru_m = {f: float(np.mean([SW[(f,sd)]['att_trust'] for sd in SEEDS])) for f in TRIG_FEATS}
DEFEAT_THRESH, VIABLE_THRESH = 0.02, 0.05

print('VERDICT, computed from the sweep above')
print(f'(defeated = defended lift <= +{DEFEAT_THRESH}; viable attack = undefended lift >= +{VIABLE_THRESH})\n')
for f in sorted(TRIG_FEATS, key=lambda f: DV[f], reverse=True):
    elim = 100*(1 - dfd_m[f]/und_m[f]) if und_m[f] > 0 else float('nan')
    print(f"  {f:<4} (d={DV[f]:.3f}): undefended {und_m[f]:+.4f} -> defended {dfd_m[f]:+.4f} +/- {dfd_s[f]:.4f}, "
          f"{elim:.1f}% of the lift removed, mean attacker trust {tru_m[f]:.4f}")

viable   = [f for f in TRIG_FEATS if und_m[f] >= VIABLE_THRESH]
nonviable= [f for f in TRIG_FEATS if und_m[f] <  VIABLE_THRESH]
broke    = [f for f in viable if dfd_m[f] > DEFEAT_THRESH]
worst_f  = max(viable, key=lambda f: dfd_m[f]) if viable else None
print()
if nonviable:
    print('Note: ' + ', '.join(f'{f} (undefended lift {und_m[f]:+.4f})' for f in nonviable) +
          ' did not carry a viable attack even with no defense, so those rows say little about the defense either way.')
if viable and not broke:
    print(f'The defense held at every trigger location that carried a viable attack. The worst residual defended')
    print(f'lift is {dfd_m[worst_f]:+.4f} +/- {dfd_s[worst_f]:.4f} ({worst_f}, d={DV[worst_f]:.3f}), under the '
          f'+{DEFEAT_THRESH} threshold, and mean attacker')
    print(f'trust never exceeds {max(tru_m[f] for f in viable):.4f}. The defense configuration was identical for '
          f'all features, so this')
    print(f'supports "trigger-agnostic" WITHIN the probed (d >= 0.05) feature set. PQP and PIP sit outside that')
    print(f'set by design and were not tested as triggers, so the paper should scope the claim to the')
    print(f'discriminative feature set wherever precision matters.')
elif broke:
    print('The evidence does NOT support "trigger-agnostic" as stated. It breaks down at: ' +
          '; '.join(f'{f} (d={DV[f]:.3f}), residual defended lift {dfd_m[f]:+.4f} +/- {dfd_s[f]:.4f} '
                    f'out of an undefended {und_m[f]:+.4f}' for f in broke))
print()
print(f'Weak-versus-strong check: defended lift at the weakest probed trigger ({weakest}, d={DV[weakest]:.3f}) is')
print(f'{dfd_m[weakest]:+.4f} +/- {dfd_s[weakest]:.4f}, versus {dfd_m[strongest]:+.4f} +/- {dfd_s[strongest]:.4f} '
      f'at the strongest ({strongest}, d={DV[strongest]:.3f});')
print(f'the difference is {dfd_m[weakest]-dfd_m[strongest]:+.4f}.')
```

**Table 6 and figure: trigger generalization at the fixed D2 configuration.** One row and one x-position per trigger feature; the defense never changes between rows. All values are mean +/- standard deviation over seeds 42, 7, and 123, with lift paired within-seed against that seed's honest baseline evaluated on the same triggered test set. "Lift eliminated" can exceed 100% because the defended lift is slightly negative.

**The headline result.** The defense holds at every trigger location. Mean defended lift is negative for all five features, from -0.0056 +/- 0.0068 (LC, the worst) to -0.0626 +/- 0.0512 (TCD, the best), while the undefended attack gains between +0.0909 (PD) and +0.3970 (LC). Both attackers are driven to effectively zero trust (at most 0.0002 in any seed) regardless of which feature carries the trigger; the trust mechanism probes all discriminative features, so moving the trigger just changes which probe slice exposes the attacker.

**The stress test the Week 8 result could not provide.** PD is the weakest feature in the probe set (d = 0.152, half the strength of the features tested before), and the defense holds there too: +0.0909 undefended falls to -0.0188 +/- 0.0150 defended. The undefended attack itself is weakest through PD, which is consistent with a low-separation feature making a less effective trigger, but it still cleared +0.05 in every seed, so the row is a real test and not a vacuous one.

**Honest caveats, stated rather than buried.** First, the seed spread at the two strongest features is wide (std 0.0638 at DO, 0.0512 at TCD), and one individual cell is positive: DO at seed 42 shows +0.0304 defended, before averaging to -0.0470 across seeds. We report the mean as the estimate, and note that a single-seed reader could have drawn either conclusion at DO; this is exactly why the sweep is multi-seed. Second, the mid-range pick was effectively a tie: the probed median d is 0.2510, exactly halfway between PC (0.2572) and LC (0.2448) since there are eight probed features, and the selection code broke the tie by list order in favor of LC. PC would have been an equally valid choice. Third, undefended lift is not monotone in d; LC carries the strongest undefended attack (+0.3970) despite its middling separation, so trigger strength and feature separation are related but not interchangeable.

**Scope.** This supports "trigger-agnostic within the discriminative (probed, d >= 0.05) feature set" at the configuration we publish. PQP and PIP are outside the probe set by design and were not tested as triggers; a trigger there would be attacking signal the detector does not use, which is a different question. The adaptive attacker limitation from Section 10 also still stands; this sweep varies the trigger location, not the attacker's strategy.
